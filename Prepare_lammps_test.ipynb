```python
from ase.io import read, write
from ase.build import make_supercell
from ase.visualize import view
import numpy as np
from pathlib import Path

folder = Path("lammps_test")

# 1. read CIF
atoms = read("TaCuN2_unrelaxed.cif")

# 2. build supercell (important for MD!)
atoms = atoms.repeat((1, 1, 1))

print("Atoms:", len(atoms))

# 3. write LAMMPS data file (optional for LAMMPS route)
write(folder / "structure.lmp", atoms, format="lammps-data")

# 4. also save restart structure
write(folder / "structure.xyz", atoms)

view(atoms)
```
