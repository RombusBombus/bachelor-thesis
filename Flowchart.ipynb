```python
from graphviz import Digraph
```

```python
dot = Digraph("FlowOtter", format="pdf")
# dot.attr(rankdir="LR", fontsize="12", labelloc="t", label="Flow-Otter")
dot.attr(rankdir="LR", fontsize="12", labelloc="t")

dot.attr("node",
         shape="box",
         style="rounded,filled",
         fontname="Helvetica")

# -------------------
# Input
# -------------------
dot.node("input", "input config\nYAML file",
         shape="ellipse",
         fillcolor="#f2f2e9")

# -------------------
# Molecular Dynamics cluster
# -------------------
with dot.subgraph(name="cluster_md") as c:
    c.attr(label="Molecular Dynamics", color="black", penwidth="2")

    c.node("md1", "LAMMPS\nFF-MD", fillcolor="#d9e6f2")
    c.node("md2", "LAMMPS\nVASP\nML-MD", fillcolor="#d9e6f2")
    c.node("md3", "LAMMPS\nMACE-MD", fillcolor="#d9e6f2")

# -------------------
# equilibration test
# -------------------
dot.node("test_eq",
         "test for\nequilibration\nand plausibility",
         fillcolor="#ead1dc")

# -------------------
# Electronic Hamiltonians (UPDATED)
# -------------------
with dot.subgraph(name="cluster_eh") as c:
    c.attr(label="Electronic Hamiltonians", color="black", penwidth="2")

    c.node("eh1", "empirical\nTB fit", fillcolor="#d9ead3")
    c.node("eh2", "Hamster\nML-H", fillcolor="#d9ead3")

    # NEW: DFT block (green)
    c.node("eh3", "DFT", fillcolor="#d9ead3")

# -------------------
# Hamiltonian test
# -------------------
dot.node("test_ham",
         "plausibility test\nof Hamiltonian",
         fillcolor="#ead1dc")

# -------------------
# Optoelectronics
# -------------------
with dot.subgraph(name="cluster_opt") as c:
    c.attr(label="Optoelectronics", color="black", penwidth="2")

    c.node("opt1", "MD + Kubo model", fillcolor="#fff2cc")
    c.node("opt2", "band gap & DOS", fillcolor="#fff2cc")

# -------------------
# Edges (Flow)
# -------------------
dot.edges([
    ("input", "md1"),
    ("input", "md2"),
    ("input", "md3"),

    ("md1", "test_eq"),
    ("md2", "test_eq"),
    ("md3", "test_eq"),

    ("test_eq", "eh1"),
    ("test_eq", "eh2"),
    ("test_eq", "eh3"),  # NEW connection to DFT

    ("eh1", "test_ham"),
    ("eh2", "test_ham"),
    ("eh3", "test_ham"),  # NEW connection

    ("test_ham", "opt1"),
    ("test_ham", "opt2"),
])

dot.render("flow_otter", view=True)
```

```python
dot = Digraph("KS_scheme")
dot.attr(rankdir="LR")
dot.attr(splines="false")
dot.attr(nodesep="1.0")
dot.attr(ranksep="1.2")

# Global node style
dot.attr("node", shape="plaintext", fontsize="24")

# Top row
dot.node("Vext", r"<<i>V</i><sub>ext</sub>(<b>r</b>)>")
dot.node("n0L", r"<<i>n</i><sub>0</sub>(<b>r</b>)>")
dot.node("n0R", r"<<i>n</i><sub>0</sub>(<b>r</b>)>")
dot.node("VKS", r"<<i>V</i><sub>KS</sub>(<b>r</b>)>")

# Bottom row
dot.node("Psi", r"&Psi;<sub>i</sub>({<b>r</b>})")
dot.node("Psi0", r"&Psi;<sub>0</sub>({<b>r</b>})")
dot.node("psi_occ", r"&psi;<sub>i=1,...,N<sub>e</sub></sub>(<b>r</b>)")
dot.node("psi", r"&psi;<sub>i</sub>(<b>r</b>)")

# Put nodes on the same ranks
with dot.subgraph() as s:
    s.attr(rank="same")
    s.node("Vext")
    s.node("n0L")
    s.node("n0R")
    s.node("VKS")

with dot.subgraph() as s:
    s.attr(rank="same")
    s.node("Psi")
    s.node("Psi0")
    s.node("psi_occ")
    s.node("psi")

# Horizontal mappings
dot.edge("Vext", "n0L",
         dir="both",
         arrowhead="normal",
         arrowtail="normal",
         label="HK")

dot.edge("n0L", "n0R",
         dir="both",
         arrowhead="normal",
         arrowtail="normal",
         label="KS")

dot.edge("n0R", "VKS",
         dir="both",
         arrowhead="normal",
         arrowtail="normal",
         label=r"HK$_0$")

# Vertical mappings
dot.edge("Vext", "Psi", arrowhead="normal")
dot.edge("n0L", "Psi0", arrowhead="normal")
dot.edge("n0R", "psi_occ", arrowhead="normal")
dot.edge("VKS", "psi", arrowhead="normal")

# Bottom-row arrows
dot.edge("Psi", "Psi0", arrowhead="normal")
dot.edge("psi", "psi_occ", arrowhead="normal")

# Render
dot.render("ks_flowchart", format="pdf", cleanup=True, view=True)
dot.render("ks_flowchart", format="png", cleanup=True)

print(dot.source)
```

```python
import graphviz

# Create a directed graph with a top-to-bottom layout
dot = graphviz.Digraph('Workflow_Diagram', comment='DFT Workflow Manager', format='pdf')

# # Global graph attributes for a clean, professional academic look
# dot.attr(rankdir='TB', size='8,11', ratio='fill')
# dot.attr(ranksep='0.01')
# dot.attr('node', fontname='Helvetica', fontsize='11', shape='box', style='filled')
# dot.attr('edge', fontname='Helvetica', fontsize='10', arrowhead='vee', arrowsize='0.7')
# 1. REMOVE ratio='fill' and size so Graphviz respects your spacing
dot.attr(rankdir='TB')

# 2. Force tight spacing (ranksep = vertical, nodesep = horizontal)
dot.attr(ranksep='0.5 consecutive') # 'consecutive' forces it tightly between layers
dot.attr(nodesep='0.2')

# 3. Tighten the node margins (internal padding) so the boxes themselves are shorter
dot.attr('node', 
         fontname='Helvetica', 
         fontsize='11', 
         shape='box', 
         style='filled',
         margin='0.1,0.05',  # Shrunk padding (Width, Height)
         height='0.1')        # Forces minimum height to be as small as text allows

dot.attr('edge', fontname='Helvetica', fontsize='10', arrowhead='vee', arrowsize='0.7')

# --- STYLE DEFINITIONS (Color Palette) ---
COLOR_MGR = '#2C3E50'      # Dark Slate for Manager
COLOR_MD = '#3498DB'       # Blue for Molecular Dynamics
COLOR_SNAP = '#E67E22'     # Orange for Snapshots
COLOR_DFT = '#2ECC71'      # Green for DFT Calculations
COLOR_PROP = '#9B59B6'     # Purple for Properties/Observables

# --- NODES ---

# Core Workflow Manager
dot.node('MGR', 'Workflow Manager\n(Automation & Orchestration)', 
         shape='plaintext', fillcolor=COLOR_MGR, fontcolor='white', penwidth='2')

# MD Trajectories Layer
dot.node('MD1', 'MD Trajectory 1\n(NVT/NPT Ensemble)', fillcolor=COLOR_MD, fontcolor='white')
dot.node('MD_DOTS', '...', shape='plaintext', fontcolor='black', fontsize='16')
dot.node('MD_N', 'MD Trajectory N\n(NVT/NPT Ensemble)', fillcolor=COLOR_MD, fontcolor='white')

# Snapshots Layer (Sampled from Trajectory 1 as prime example)
dot.node('S1', 'Snapshot 1', fillcolor=COLOR_SNAP, fontcolor='white')
dot.node('S2', 'Snapshot 2', fillcolor=COLOR_SNAP, fontcolor='white')
dot.node('S_DOTS', '...', shape='plaintext', fontcolor='black', fontsize='16')
dot.node('S_M', 'Snapshot M', fillcolor=COLOR_SNAP, fontcolor='white')

# DFT Calculation Node
dot.node('DFT', 'Aggregation over\nSnapshots', 
         shape='hexagonal', fillcolor=COLOR_DFT, fontcolor='white', penwidth='1.5')

# Observables / Properties Layer
dot.node('BG', 'Band Gap', fillcolor=COLOR_PROP, fontcolor='white')
dot.node('DOS', 'Density of States\n(DOS)', fillcolor=COLOR_PROP, fontcolor='white')
# dot.node('OTHER', 'Other Observables\n(Total Energy, Charge)', fillcolor=COLOR_PROP, fontcolor='white')


# --- EDGES (Connections) ---

# Manager to MD Trajectories
dot.edge('MGR', 'MD1')
dot.edge('MGR', 'MD_DOTS', style='dotted', arrowhead='none')
dot.edge('MGR', 'MD_N')

# MD Trajectory to Snapshots
dot.edge('MD1', 'S1')
dot.edge('MD1', 'S2')
dot.edge('MD1', 'S_DOTS', style='dotted', arrowhead='none')
dot.edge('MD1', 'S_M')

# Snapshots to DFT (Using a subgraph or clean convergence to show it applies to all)
# To keep the graph clean, we route the sampled snapshots into the DFT engine
dot.edge('S1', 'DFT')
dot.edge('S2', 'DFT')
dot.edge('S_M', 'DFT')

# DFT to Properties
dot.edge('DFT', 'BG')
dot.edge('DFT', 'DOS')
# dot.edge('DFT', 'OTHER')


# --- GROUPING (Optional clusters to keep ranks aligned visually) ---
# This ensures snapshots stay on the same horizontal level
with dot.subgraph() as s:
    s.attr(rank='same')
    s.node('S1')
    s.node('S2')
    s.node('S_M')

# render graph as pdf
dot.render('Workflow_Diagram', view=True)
```
